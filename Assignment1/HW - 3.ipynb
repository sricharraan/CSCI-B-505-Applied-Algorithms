Calculating the sum, product and the reciprocal sum


```python
def f(*x, **y):
    
    def sum1(x):
        s = 0
        if x:
            for i in x:
                s += i   # Calculating the sum of the numbers in the list
        return s

    def prod1(x):
        s = 1
        if x:
            for i in x:
                s *= i  # Calculating the product of the numbers in the list
        return s

    def reciprocal(x): # Calculating the reciprocal sum of the numbers in the list
        s = 0
        x = [i() for i in (lambda num = num:1/num for num in x)] # Using lambda, we calculate the sum of reciprocals
        return sum1(x) #Call the sum function
    

    if y["action"] == "sum":
        return sum1(*x)
    elif y["action"] == "prod":
        return prod1(*x)
    elif y["action"] == "reciprocal sum":
        return reciprocal(*x)
    else:
        return f"bad argument: {y}"


if __name__ == '__main__':
    
    xlst = [1,2,3,4,5]
    print("Sum : {}".format(f(xlst, action = "sum")))
    print("Product : {}".format(f(xlst, action = "prod")))
    print("Reciprocal sum : {}".format(f(xlst, action = "reciprocal sum")))
```

In the example already given, the addition operator was overloaded to make the complex addition similar to the native type of addition.

Similarly, we write a cmul function for the multiplication of complex integers.


```python
class complex_:
    def __init__(self, re=0, im=0):
        self.re = re
        self.im = im

    def get_re(self):
        return self.re

    def get_im(self):
        return self.im

    def __str__(self):
        g = lambda x:"+" if x >= 0 else ""
        return f"({self.re}{g(self.im)}{self.im}i)"

    def cadd(self, other):
        new_re = self.get_re() + other.get_re()
        new_im = self.get_im() + other.get_im()
        return complex_(new_re, new_im)

    def __add__(self, other):
        new_re = self.get_re() + other.get_re()
        new_im = self.get_im() + other.get_im()
        return complex_(new_re, new_im)

    def cmul(self, other):
        new_re = (self.get_re() * other.get_re())- (self.get_im() * other.get_im())
        new_im = (self.get_re() * other.get_im())+ (self.get_im() * other.get_re())
        return complex_(new_re, new_im)

    def __mul__(self, other):
        new_re = (self.get_re() * other.get_re())- (self.get_im() * other.get_im())
        new_im = (self.get_re() * other.get_im())+ (self.get_im() * other.get_re())
        return complex_(new_re, new_im)


if __name__ == '__main__':

    w = complex_(1,-3)
    x = complex_(-1,3)
    y = complex_(1,3)
    z = complex_(-1,-3)
    print("Input values 1:{}" .format(w))
    print("Input values 2:{}" .format(x))
    print("Input values 3:{}" .format(y))
    print("Input values 4:{}" .format(z))

    print("Sum using function from class complex:{}".format(w.cadd(x).cadd(y).cadd(z)))
    print("Multiplication using function from class complex :{}".format(w.cmul(x).cmul(y).cmul(z)))
    print("Sum just putting values in print() :{}".format((1-3j) + (-1+3j) + (1+3j) + (-1-3j)))
    print("Multiplication just putting values in print() :{}".format((1-3j) * (-1+3j) * (1+3j) * (-1-3j)))
    print("Sum using variables :{}".format(w + x + y + z))
    print("Multiplication using variables :{}".format(w * x * y * z))
```


## Question - 4

We are taking a list 'I' here and appending it with the value from 1 to 2 power 10.

In the list S1, we append random integers using randint() function for the 3 algorithms.

For the three algorithms given, we take three lists - end1, end2, end3.

We append the running time of each algorithm to the respective lists.

Using log-log scale for the graph, we plot the input size vs run time of each algorithm in the graph using matplotlib.


```python
#Algorithm - 1
#Time complexity - O(n^2)
def average1(S):
    n = len(S)
    my_average = [0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += S[i]
        my_average[j] = total / (j+1)
    return my_average
```

```python
#Algorithm - 2
#Time complexity - O(n)
def average2(S):
    n = len(S)
    my_average= [0] * n
    for j in range(n):
        my_average[j] = sum(S[0:j+1]) / (j+1)
    return my_average
```

```python
#Algorithm - 3
#Time complexity - O(n)
def average3(S):
    n = len(S)
    my_average = [0] * n
    total = 0
    for j in range(n):
        total += S[j]
        my_average[j] = total / (j+1)
    return my_average
```

```python
import random as r
import matplotlib.pyplot as plt
from timeit import default_timer 

I = [] #Input Size

for i in range(pow(2,10)):
    I.append(i)
end1 = []  #Runtime for Algorithm - 1
end2 = []  #Runtime for Algorithm - 2
end3 = []  #Runtime for Algorithm - 3
S1 = [] 

for j in I:
    for i in range(j):
        S1.append(r.randint(0,4000))
        
    #Algorithm - 1
    start1 = default_timer()
    avg = average1(S1)
    end1.append((default_timer()-start1))
    #Algorithm - 2
    start2 = default_timer()
    avg = average2(S1)
    end2.append((default_timer()-start2))
    #Algorithm - 3
    start3 = default_timer()
    avg = average3(S1)
    end3.append((default_timer()-start3))
    S1.clear()

#Plotting the log-log graph   
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Input I')
plt.ylabel('Running time per Input size')
plt.title('Running time visualization of each algorithm')
plt.plot(I,end1, color = 'b', label = 'average1 function')
plt.plot(I,end2, color = 'r', label = 'average2 function')
plt.plot(I,end3, color = 'g', label = 'average3 function')
plt.legend()
```

## Question - 5

According to our understanding each algorithm returns true if there are unique elements in the list else it will return false.

For the largest value of input size, the values in the list should be unique i.e. worst case scenario. 

Since, randint() function will generate random numbers and there is a possibility of two numbers getting repeated.

Therefore, the value of list 'S' will have values from 1 to the input size which will make the list unique.

In algorithm1 we can clearly see that the time complexity is O(n^2) and therefore after trying for values 1-10000 we saw that the run time was less than 0.01 sec therefore we started appending values from 10000 to 50000 with a step size of 5000. 

Once we have calculated the run time of algorithm1, we have plotted a Running time visualization graph for the same and have used interp() function(in numpy class) to find the most accurate largest value of input size.

Our final answer i.e. The largest input size for algorithm1 whose runtime is less than 45 sec is 33592.

```python
def algorithm1(S):
    #S:sequence
    for j in range(len(S)):
        for k in range(j+1, len(S)):
             if S[j] == S[k]:
                    return False
    return True
```

```python
import matplotlib.pyplot as plt
from timeit import default_timer 
import numpy as np
Is = []
If1 = []
S = []
end1  = []
j = 1
for i in range(10000,50000,5000):
    Is.append(i)
#print(Is)
while (j>0):
     for i in Is:
            for k in range(i):
                S.append(k)
            start1 = default_timer()
            algo1 = algorithm1(S)
            end1.append(default_timer()-start1)
            If1.append(i)
            #print(If1)
            #print(end1)
            if (end1[-1] > 45):
                ans = If1[-2]
                best_time = end1[-2]
                j = 0
                break
            S.clear()
print("The largest input size is {} and the running time for the same is {}".format(ans,best_time))
plt.xlabel('Input I')
plt.ylabel('Running time per Input size')
plt.title('Running time visualization of algorithm1')
plt.plot(If1,end1, color = 'b')
print("The largest input size with accuracy (from the graph): {}".format(int(np.interp(45, end1, If1))))
```

```python
def algorithm3(S, start, stop):
    if stop - start <= 1: return True
    elif not algorithm3(S, start, stop-1): return False
    elif not algorithm3(S, start+1, stop): return False
    else: return S[start] != S[stop-1]
```
